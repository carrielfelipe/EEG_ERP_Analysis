```python
from collections import Counter

# Obtener la lista de eventos y sus nombres
eventos = Epocas['Sujeto_1'].events  # Supone que eventos es una matriz (con tiempos, códigos, y otros datos)
eventos_tiempos = eventos[:, 0]  # Asumiendo que la primera columna es el tiempo
eventos_codigos = eventos[:, 2]  # Asumiendo que la tercera columna es el código del evento

# Obtener los nombres de los eventos
eventos_nombres = [nombre for codigo in eventos_codigos for nombre, valor in Epocas['Sujeto_1'].event_id.items() if codigo == valor]

# Contar la frecuencia de cada marca y almacenar las posiciones y tiempos
conteo_marcas = Counter(eventos_nombres)
posiciones_marcas = {marca: [] for marca in conteo_marcas.keys()}
tiempos_marcas = {marca: [] for marca in conteo_marcas.keys()}

# Almacenar las posiciones y tiempos de cada marca
for idx, (nombre, tiempo) in enumerate(zip(eventos_nombres, eventos_tiempos)):
    posiciones_marcas[nombre].append(idx)
    tiempos_marcas[nombre].append(tiempo)

# Mostrar los resultados con tiempos
print("Frecuencia de cada marca, posiciones y tiempos:")
for marca, frecuencia in conteo_marcas.items():
    print(f"{marca}: {frecuencia} veces, posiciones: {posiciones_marcas[marca]}, tiempos: {tiempos_marcas[marca]}")
```

```python
import numpy as np

# Definir los pools con los nombres de los canales
pools = {
    "FrontoCentral": ['C1', 'C2', 'Cz', 'FC1', 'FC2', 'FCz'],
    "Fronto": ['F1', 'F2', 'FC1', 'FC2', 'FCz', 'Fz'],
    "Frontal": ['F1', 'F2', 'FCz', 'Fz'],
    "Cz": ['Cz'],
    "Parietal": ['POz', 'Pz'],
    "Occipital": ['Iz', 'O1', 'O2', 'Oz']
}

# Inicializa el diccionario para guardar los datos agrupados por pool
PooledData = {}

# Recorre cada sujeto en el diccionario Epocas
for sujeto, epochs in Epocas.items():
    if epochs is None:
        # Si no hay épocas, guarda None para este sujeto
        PooledData[sujeto] = None
        print(f"{sujeto} no tiene datos de épocas.")
        continue

    # Extrae los datos de las épocas
    epochs_data = epochs.get_data()  # Forma: (n_epocas, n_canales, n_muestras)
    channel_names = epochs.info['ch_names']

    # Crear los arreglos para cada pool
    pools_data = {}
    for pool_name, pool_channels in pools.items():
        # Obtener los índices de los canales correspondientes
        channel_indices = [channel_names.index(ch) for ch in pool_channels if ch in channel_names]
        if not channel_indices:
            print(f"Advertencia: No se encontraron canales para el pool {pool_name} en {sujeto}.")
            continue

        # Extraer los datos de esos canales
        pools_data[pool_name] = epochs_data[:, channel_indices, :]

    # Promediar los datos dentro de cada pool
    pooled_data = {}
    for pool_name, data in pools_data.items():
        # Promediar a lo largo del eje de los canales (axis=1)
        pooled_data[pool_name] = data.mean(axis=1)

    # Guardar los datos promediados en el diccionario final
    PooledData[sujeto] = pooled_data
    #print(f"Pooling completado para {sujeto}.")

# Ahora el diccionario PooledData contiene los datos agrupados o None si no hay datos
```

# Antes del Artifact Rejection

```python
import matplotlib.pyplot as plt
import numpy as np

# Frecuencia de muestreo (en Hz)
sampling_rate = 1024.0

# Determinar el número de sujetos (40 sujetos fijos)
total_subjects = 40

# Configurar la cuadrícula de subgráficos (5 columnas y 8 filas)
cols = 5
rows = 8
fig, axes = plt.subplots(rows, cols, figsize=(20, rows * 4), squeeze=False)

# Iterar sobre los sujetos
for idx in range(1, total_subjects + 1):
    # Determinar la posición del subgráfico
    row, col = divmod(idx - 1, cols)
    ax = axes[row, col]

    # Identificar el sujeto actual
    sujeto = f"Sujeto_{idx}"

    # Verificar si el sujeto tiene datos en el pool Frontal
    if sujeto in PooledData and PooledData[sujeto] is not None and "Frontal" in PooledData[sujeto]:
        # Extraer los datos del pool Frontal
        frontal_data = PooledData[sujeto]["Frontal"]

        # Crear un vector de tiempo para el eje X
        n_muestras = frontal_data.shape[1]
        time = np.arange(n_muestras) / sampling_rate

        # Superponer todas las épocas del pool Frontal (en gris)
        for epoch in frontal_data:
            ax.plot(time, epoch, color='gray', alpha=0.5)

        # Calcular y graficar el promedio de las épocas (en naranja)
        mean_epoch = frontal_data.mean(axis=0)
        ax.plot(time, mean_epoch, color='orange', linewidth=2, label='Promedio')

        # Mejoras del subgráfico
        ax.set_title(f'{sujeto} - Frontal')
        ax.set_xlabel('Tiempo (s)')
        ax.set_ylabel('Amplitud')
        ax.grid(True)
        ax.legend(loc='upper right')
    else:
        # Si no hay datos, dejar el gráfico en blanco con un título
        ax.set_title(f'{sujeto} - Sin datos')
        ax.axis('off')

# Ajustar el diseño y mostrar la figura
plt.tight_layout()
plt.show()
```
