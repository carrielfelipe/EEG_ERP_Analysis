```python
import numpy as np
import mne

# Cargar info de electrodos (usamos el sistema estndar de 10-20)
montage = mne.channels.make_standard_montage("standard_1020")

# Crear info de un EEG ficticio con 32 canales y frecuencia de muestreo de 100 Hz
info = mne.create_info(ch_names=montage.ch_names, sfreq=100, ch_types="eeg")
info.set_montage(montage)

# Crear datos simulados (valores aleatorios en cada electrodo)
data = np.random.randn(len(montage.ch_names))  # Un valor por electrodo

# Obtener posiciones de los electrodos en 2D
pos = np.array([info.get_montage().get_positions()['ch_pos'][ch][:2] for ch in info.ch_names])

# Graficar el topomap
mne.viz.plot_topomap(data, pos, cmap="coolwarm", show=True)
```

```python
import numpy as np
import mne

# Definir la cantidad de electrodos y generar datos ficticios
n_channels = 64
fake_eeg_data = np.random.randn(n_channels)  # Seal aleatoria por canal

# Crear una info con la disposicin estndar 10-20
montage = mne.channels.make_standard_montage("standard_1020")
info = mne.create_info(ch_names=montage.ch_names[:n_channels], sfreq=250, ch_types="eeg")
info.set_montage(montage)

# Crear un objeto Evoked con los datos ficticios
evoked = mne.EvokedArray(fake_eeg_data[:, np.newaxis], info)

# Graficar el topoplot
mne.viz.plot_topomap(evoked.data[:, 0], evoked.info, cmap="RdBu_r", contours=10)
```

```python
import numpy as np
import mne
import matplotlib.pyplot as plt

# Definir la cantidad de electrodos y generar datos ficticios
n_channels = 64
fake_eeg_data = np.random.randn(n_channels)  # Seal aleatoria por canal

# Crear una info con la disposicin estndar 10-20
montage = mne.channels.make_standard_montage("standard_1020")
info = mne.create_info(ch_names=montage.ch_names[:n_channels], sfreq=250, ch_types="eeg")
info.set_montage(montage)

# Crear un objeto Evoked con los datos ficticios
evoked = mne.EvokedArray(fake_eeg_data[:, np.newaxis], info)

# Usar el tamao de la figura dentro de `plot_topomap`
fig, ax = plt.subplots(figsize=(10, 8))  # Ajustar el tamao de la figura

# Graficar el topoplot con la barra de color
im, _ = mne.viz.plot_topomap(evoked.data[:, 0], evoked.info, cmap="RdBu_r", contours=10, axes=ax, show=False)

# Agregar barra de color
fig.colorbar(im, ax=ax, orientation='vertical', label='Valor de la seal')

plt.show()
```
